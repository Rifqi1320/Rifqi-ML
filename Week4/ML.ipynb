```python
# Importing Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
```

**Scatter Plots**

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
# Recover default matplotlib settings
mpl.rcParams.update(mpl.rcParamsDefault)
%matplotlib inline
```

```python
plt.figure(figsize=(7,7))
sns.scatterplot(x=df.Age , y=df.Fare , hue=df.Sex)
plt.show()
```

**Relplot**

```python
insu = pd.read_csv("insurance.csv")
insu.head()
```

```python
sns.set_style("white") 

sns.set(rc={'xtick.labelsize':12,'ytick.labelsize':12,'axes.labelsize':12  , "axes.grid":False})
sns.relplot(x="bmi", y="charges", hue="sex",
                col="sex", row="region", data=insu)
```

**Bar Plot**

```python
insu.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=insu.sex , y=insu.bmi , hue=insu.smoker)
sns.despine()
plt.show()
```

**Count Plot**

```python
pokemon = pd.read_csv("pokemon.csv")
pokemon.head()
```

```python
sns.catplot(y = "Type 1", kind = "count",data = pokemon)
plt.show()
```

**Violin Plot**

```python
pokemon.head()
```

```python
plt.figure(figsize=(7,7))
sns.catplot(x="Attack" , y = "Type 1" ,kind="violin" ,data=pokemon)
plt.show()
```

**Dist Plot**

```python
pokemon.head()
```

```python
plt.figure(figsize=(11,7))
sns.distplot( pokemon["Defense"] )
plt.show()
```

**KDE Plot**

```python
insu.head()
```

```python
plt.figure(figsize=(6,8))
sns.kdeplot(insu.bmi,insu.charges,shade=True,cmap="Reds", shade_lowest=False)
plt.show()
```

**Swarm Plot**

```python
insu.head()
```

```python
plt.figure(figsize=(7,7))
sns.swarmplot(x=insu.charges , y=insu.smoker)
plt.show()
```

**Strip Plot**

```python
insu.head()
```

```python
plt.figure(figsize=(16,8))
sns.stripplot(x=insu.sex ,palette="Set1", y = insu.bmi , linewidth=1 , jitter=1)
plt.show()
```

**Box Plot**

```python
insu.head()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(8,6))
sns.boxplot(x= insu.smoker , y= insu.charges )
```
